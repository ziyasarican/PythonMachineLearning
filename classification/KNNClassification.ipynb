```python
# TenYearCHD'leri tahmin etme
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("framingham.csv")
df.head()
```

```python
# Önce datayı tanımak için özet tabloyu oluşturalım
df.describe()
```

```python
# Eksik veri var mı bakalım
df.isnull().sum()
```

```python
# Eksik veriler sütunların ortalaması ile doldurulup, değişimi inplace=true ile kalıcı yapılır
df.fillna(df.mean(), inplace=True)
df.loc[:,"education"].mean()
```

```python
# Artık eksik veri yok
df.isnull().sum()
```

```python
# y=wx+b

y = df.loc[:,"TenYearCHD"]
x = df.drop(["TenYearCHD"], axis=1)
```

```python
# train-test datalarını oluşturma
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 0)
```

```python
# kullanacağımız modeli ekleme
from sklearn.neighbors import KNeighborsClassifier
# n_neighbors: bakılacak komşu sayısı. metric: mesafeyi hangi modelle ölçüleceği
KNN = KNeighborsClassifier(n_neighbors=5, metric="minkowski")
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred = KNN.predict(x_test)
y_pred
```

```python
KNN.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix

# Confusion Matrix
confusion_matrix(y_test, y_pred)
```

```python
plot_confusion_matrix(KNN, x_test, y_test, cmap = "YlGnBu")
```
