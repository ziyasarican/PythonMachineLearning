- Ensemble Learning: Birden fazla sınıflandırma/tahmin algoritması aynı anda kullanılarak daha başarılı bir sonuç bulunabilir.
- Random Forest'da bunun bir örneğidir.

image.png

- Training aşamasında bütün verileri tek bir decision-tree için kullanmayıp parça parça alır. Bu da birden fazla decision-tree oluşması anlamına gelir.
- Yani train verileri birden fazla decision-tree için train edilir.
---
- Bu oluşan decision-tree'ler alınıp majority-vote(çoğunluğun oyu)'na tabi tutulur.
- Yani decision-tree'lerin bir düğüm noktasında çoğunluğun sağlandığı dal takip edilir.
- Sınıflandırmadaki yaklaşım: Çoğunluk erkek diyorsa erkek, kadın diyorsa kadın.
---
- Tahmindeki yaklaşım: Decision-tree'lerdeki bulunan değerlerin ortalaması alınır.
- Örneği boy'u bulmaya çalışıyoruz, farklı decision-tree'lerdeki bulunan boy'ların ortalaması alınır.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("maaslar.csv")

x = df.loc[:,["Egitim Seviyesi"]]
y = df.loc[:,["maas"]]
```

```python
from sklearn.ensemble import RandomForestRegressor

# n_estimators-kaç tane decision tree çizileceği
randForestReg = RandomForestRegressor(n_estimators=10, random_state=0)

randForestReg.fit(x,y)
```

```python
# decision-tree'de tablodaki verilerden farklı değer üretilemezdi
# random-forest ile tablodakiler dışında bir değer üretmek mümkün
# bu daha başarılı sonuçlar verebilir
randForestReg.predict([[5.5]])
```

```python
plt.scatter(x,y, color="red")
plt.plot(x,randForestReg.predict(x))
```
