- Training aamasnda btn verileri tek bir decision-tree iin kullanmayp para para alr. Bu da birden fazla decision-tree olumas anlamna gelir.
- Yani train verileri birden fazla decision-tree iin train edilir.

- Bu oluan decision-tree'ler alnp majority-vote(ounluun oyu)'na tabi tutulur.
- Yani decision-tree'lerin bir dm noktasnda ounluun saland dal takip edilir.
- Snflandrmadaki yaklam: ounluk erkek diyorsa erkek, kadn diyorsa kadn.
---
- Tahmindeki yaklam: Decision-tree'lerdeki bulunan deerlerin ortalamas alnr.
- rnei boy'u bulmaya alyoruz, farkl decision-tree'lerdeki bulunan boy'larn ortalamas alnr.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("maaslar.csv")

x = df.loc[:,["Egitim Seviyesi"]]
y = df.loc[:,["maas"]]
```

```python
from sklearn.ensemble import RandomForestRegressor

# n_estimators-ka tane decision tree izilecei
randForestReg = RandomForestRegressor(n_estimators=10, random_state=0)

randForestReg.fit(x,y)
```

```python
# decision-tree'de tablodaki verilerden farkl deer retilemezdi
# random-forest ile tablodakiler dnda bir deer retmek mmkn
# bu daha baarl sonular verebilir
randForestReg.predict([[5.5]])
```

```python
plt.scatter(x,y, color="red")
plt.plot(x,randForestReg.predict(x))
```
