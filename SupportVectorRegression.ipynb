```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("maaslar.csv")
```

```python
# polinomial regression ile kullanarak karşılaştırması

x = df.loc[:,["Egitim Seviyesi"]]
y = df.loc[:,["maas"]]
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures
pol_reg = PolynomialFeatures(degree=4)

x_pol = pol_reg.fit_transform(x)
y_pol = pol_reg.fit_transform(y)

lr = LinearRegression()
lr.fit(x_pol,y)

plt.scatter(x,y)
plt.plot(x,lr.predict(pol_reg.fit_transform(x)))
```

```python
# x,y değerlerini scale ediyoruz
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_scaling = sc.fit_transform(x)
y_scaling = sc.fit_transform(y)
```

```python
from sklearn.svm import SVR

# kernel: hangi fonksiyon türü kullanılacak
svr_reg = SVR(kernel="rbf",)
svr_reg.fit(x_scaling,y_scaling)

plt.scatter(x_scaling, y_scaling, color="green")
plt.plot(x_scaling,svr_reg.predict(x_scaling))
```
