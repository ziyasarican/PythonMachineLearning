Amaç veriler için en uygun doğruyu çizmektir.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("satislar.csv")
df
```

```python
from sklearn.model_selection import train_test_split

# y = w*x + b

y = df[["Satislar"]]
x = df[["Aylar"]]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
# scaling işlemi
# veriler standartlaştırılıyor, nöronların daha kolay ilişki kurabileceği seviyeye getiriyor
# bunun için farklı kütüphaneler mevcut, araba fiyatları örneğinde başka kütüphane kullanmıştım

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

y_train = sc.fit_transform(y_train)
y_test = sc.fit_transform(y_test)
```

```python
# verilerin öğrenilmesi için model oluşturup çağırıyoruz
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
# test verilerini modele uyguluyoruz
# modelden, test verilerinden sonuç üretmesini istiyoruz
predict = lr.predict(x_test)
predict.squeeze()
```

```python
plt.scatter(x_train, y_train, color = "green")
plt.plot(x_test, predict)
```
