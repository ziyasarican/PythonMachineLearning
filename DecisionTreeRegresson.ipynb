```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("maaslar.csv")

x = df.loc[:,["Egitim Seviyesi"]]
y = df.loc[:,["maas"]]
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures
pol_reg = PolynomialFeatures(degree=4)

x_pol = pol_reg.fit_transform(x)
y_pol = pol_reg.fit_transform(y)
```

```python
from sklearn.tree import DecisionTreeRegressor
decTreeReg = DecisionTreeRegressor(random_state=0)

decTreeReg.fit(x,y)

plt.scatter(x,y,color="red")
plt.plot(x,decTreeReg.predict(x))
```

```python
# model iin uygun grnse de baka tahminlerde karar aac yaps sunduu iin olduka yanl sonular verebilir
# yani bulmu olduu aralktaki deeri dndrr
decTreeReg.predict([[12]])
```
